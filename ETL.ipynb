# ETL VICTIMAS

```python
victimas = pd.read_excel('data/Homicidios.xlsx', sheet_name='VICTIMAS')
```

```python
victimas.info()
```

```python
victimas.head(5)
```

```python
victimas ["EDAD"] = pd.to_numeric(victimas["EDAD"], errors="coerce")
victimas ["EDAD"] = victimas["EDAD"].astype("Int64")
```

```python
victimas ["EDAD"].unique()
```

```python
victimas["EDAD"] = victimas["EDAD"].fillna(0.0)
```

```python
victimas ["EDAD"].unique()
```

```python
victimas.isnull().sum()
```

```python
victimas.columns = [x.capitalize() for x in victimas.columns]
```

Personalizar una columna para determinar el rango etario

```python
# Definir los rangos etarios
bins = [0, 20, 40, 60, 80, 100]
labels = ['Menos de 20', ' Entre 21- 40', 'Entre 41-60', 'Entre 61-80', 'Más de 80']

# Utilizar pd.cut()
victimas['Rango_etario'] = pd.cut(victimas['Edad'], bins=bins, labels=labels, right=False)
```

```python
victimas.head(5)
```

Realizo un merge de los dos dataframes. Pero tengo que modificar el titulo del id de victimas.

```python
victimas = victimas.rename(columns={'Id_hecho': 'Id'})
```

```python
victimas.head(0)
```

```python
hechos.head(0)
```

```python
victimas_hechos= victimas.merge(hechos, on="Id", how="left")
```

```python
victimas_hechos.head(5)
```

```python
victimas_hechos.to_csv("data/unificado.csv", index=False)
```
